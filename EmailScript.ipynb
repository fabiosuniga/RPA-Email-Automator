```python
import pyautogui
import time
```

```python
time.sleep(6)

# Seleciona o arquivo na área de trabalho
pyautogui.moveTo(x=1630, y=237, duration=1)
time.sleep(1)
pyautogui.click()
time.sleep(1)

# Copia o arquivo
pyautogui.hotkey("ctrl", "c")
time.sleep(1)

# Abre o navegador
pyautogui.moveTo(x=35, y=549, duration=1)
time.sleep(1)
pyautogui.doubleClick()
time.sleep(3)

# Maximiza a janela
pyautogui.hotkey("win", "up")
time.sleep(2)

# Digita o gmail na barra de endereço
pyautogui.hotkey("ctrl", "l")
time.sleep(1)
pyautogui.typewrite("gmail.com", interval=0.1)
time.sleep(1)
pyautogui.press("enter")
time.sleep(6)

# Clicar em "novo email"
pyautogui.moveTo(x=74, y=215, duration=1)
pyautogui.click()
time.sleep(2)

# Digita o destinatário
pyautogui.typewrite("fabiosunigaborazio@gmail.com", interval=0.1)
time.sleep(1)
pyautogui.press("enter")
time.sleep(1)
pyautogui.press("tab")
time.sleep(1)

# Digita o assunto
pyautogui.typewrite("Atividade para entrega" , interval=0.1)
time.sleep(1)
pyautogui.press("tab")
time.sleep(1)

# Cola o arquivo
pyautogui.hotkey("ctrl", "v")
time.sleep(5)

# Envia o email
pyautogui.moveTo(x=1315, y=1004, duration = 1)
time.sleep(2)
pyautogui.click()
```
